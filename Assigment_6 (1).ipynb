###***1. What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making ?***

In hypothesis testing, we make decisions based on sample data — either to reject or fail to reject a null hypothesis (H₀). However, because we rely on samples (not the entire population), there’s always a chance of making a wrong decision. These wrong decisions are called Type I and Type II errors.

###***Type I Error -***

* ***Definition :*** Rejecting the null hypothesis when it is actually true.

You think there’s an effect or difference when there isn’t one.

* ***Probability :*** Denoted by α (alpha) — this is the significance level you set .

***Example -***

Suppose a medical test says a patient has a disease, but they actually don’t.

→ The test made a Type I error.


###***Impact :***

* Leads to false alarms.

* In science, it may mean publishing a “discovery” that isn’t real.

* In business, it could mean acting on a marketing strategy that doesn’t actually improve sales.


###***Type II Error -***


* ***Definiton :*** Failing to reject the null hypothesis when it is actually false.

You miss a real effect or difference.

* ***Probability :***

Denoted by β (beta).

Power of a test = 1 − β (the probability of correctly detecting a true effect).


***Example -***

A medical test fails to detect a disease that a patient does have.
→ The test made a Type II error.

###***Impact :***

* Leads to missed opportunities.

* In research, it could mean overlooking a real effect.

* In quality control, it could mean letting defective products pass inspection .

###***2. What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis ?***

The p-value is one of the most important (and often misunderstood) concepts in hypothesis testing.

###***Definition -***

The p-value is the probability of obtaining a sample result (or one more extreme) if the null hypothesis (H₀) is actually true.

###***Example -***

Suppose we’re testing whether a new drug lowers blood pressure more than a placebo.

* H₀: The drug has no effect (mean difference = 0) .

* H₁: The drug lowers blood pressure (mean difference < 0) .

* α = 0.05

After analyzing data, you get p = 0.02.




* If the drug truly had no effect, there’s only a 2% chance of getting results as extreme as what we observed.

* Since 0.02 < 0.05, we reject H₀ → conclude that the drug likely does have an effect.


###***3. Explain the difference between a Z-test and a T-test, including when to use each.***

The Z-test and T-test are both statistical methods used in hypothesis testing to determine whether there is a significant difference between sample data and a population parameter (or between two sample means).

***(I)  Z-Test -***

* The population standard deviation (σ) is known.

* The sample size (n) is large (typically n ≥ 30), so the sampling distribution of the mean is approximately normal.


###***Example -***

* Testing whether the mean height of a population is 170 cm when σ = 10 cm is known.

* Comparing a large sample mean to a known population mean.


***Distribution used:***

Standard Normal Distribution (Z-distribution).

***(II) T-Test -***


* The population standard deviation (σ) is unknown.

* The sample size (n) is small (typically n < 30).

* The population is approximately normally distributed.


###***Examples -***

* Testing whether a small sample of students’ test scores differs from a known average, when σ is unknown.

* Comparing the means of two small samples (independent or paired).

***Distribution used:***

Student’s t-distribution, which has heavier tails (more variability) than the Z-distribution, especially for small n.
As n increases, the t-distribution approaches the normal (Z) distribution.

###***4. What is a confidence interval, and how does the margin of error influence its width and interpretation?***

A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter (such as the population mean or proportion) with a specified level of confidence (e.g., 90%, 95%, or 99%).

Confidence Interval= Point Estimate ± Margin of Error

***Example -***

if you estimate that the average test score is 75 with a 95% CI of ±3, then the interval is:

[72,78]

This means you are 95% confident that the true population mean lies between 72 and 78.

###***Mrgin of Error -***

The margin of error (MOE) is the amount added and subtracted from the point estimate to create the interval. It reflects the uncertainty in the estimate.

For a mean, a simplified formula is:

Margin of Error = Critical Value × Standard Error

* The critical value comes from a distribution (Z or t), depending on sample size and known/unknown population variance.

* The standard error measures variability in the sampling process.

The margin of error directly determines how wide or narrow the interval is:

***Larger margin of error → Wider CI***

Less precision, more uncertainty.

***Smaller margin of error → Narrower CI***

More precision, less uncertainty.


###**Interpretation -**

Say we compute a 95% CI for average weight:

65±2

So the interval is

[63,67]. Proper interpretation:

We are 95% confident that the true mean weight of the population lies between 63 and 67 kg.

Not correct to say: "There is a 95% probability the true mean is between 63 and 67."

The true value is fixed; the probability refers to the method, not the parameter.

###***5. Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?***

ANOVA (Analysis of Variance) is a statistical method used to determine whether there are significant differences between the means of three or more independent groups.

If you compared three or more groups using multiple t-tests, the probability of making a Type I error (false positive) increases.

ANOVA solves this by testing all groups simultaneously with one overall hypothesis test:

***Null hypothesis (H₀)***

All group means are equal
  
  μ1 ​= μ2 ​= μ3 ​= ⋯ = μk​

***Alternative hypothesis (Hₐ)***

At least one mean is different
(not necessarily all, just one or more)

###***6. Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.***

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = np.array([72, 68, 75, 70, 74, 69, 71, 73, 72, 70])

# Given population values
population_mean = 70
population_std = 4

# Sample statistics
sample_mean = np.mean(data)
sample_size = len(data)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print results
print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample provides sufficient evidence that the population mean is different from 70.")
else:
    print("Fail to reject the null hypothesis: The sample does not provide enough evidence to conclude the population mean is different from 70.")
```

###***7. Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using NumPy and plot the histogram.***

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate binomial dataset
n = 10
p = 0.5
sample_size = 1000
data = np.random.binomial(n, p, sample_size)

# Plot histogram
plt.figure()
plt.hist(data, bins=range(0, n+2), edgecolor='black')
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

###***8. Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.***

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
num_samples = 5000
sample_size = 50

# Generate multiple samples from a non-normal distribution (exponential)
sample_means = []
for _ in range(num_samples):
    sample = np.random.exponential(scale=1.0, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.figure()
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("CLT Demonstration: Sampling Distribution of Means\n(Exponential Distribution, n=50)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

###***9. Write a Python function to calculate and visualize the confidence interval for a sample mean.***

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence=0.95):
    """
    Calculates and visualizes the confidence interval for a sample mean using a t-distribution.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default = 0.95)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    # t-critical value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of Error
    margin_of_error = t_critical * std_err

    # Confidence Interval
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    # Print results
    print(f"Sample Mean = {mean:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: [{lower_bound:.4f}, {upper_bound:.4f}]")

    # Plot visualization
    plt.figure()
    plt.errorbar(x=[0], y=[mean], yerr=margin_of_error, fmt='o', capsize=10)
    plt.title(f"Mean with {int(confidence*100)}% Confidence Interval")
    plt.ylabel("Value")
    plt.xticks([])
    plt.show()


# Example usage with sample data
sample_data = np.random.normal(loc=50, scale=10, size=30)
plot_confidence_interval(sample_data, confidence=0.95)
```

###***10. Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.***

```python
import numpy as np
from scipy.stats import chisquare

# Observed and expected frequencies
observed = np.array([25, 30, 20, 25])  # observed counts
expected = np.array([25, 25, 25, 25])  # expected equal distribution

# Perform Chi-square Goodness-of-Fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("p-value:", p_value)
```
